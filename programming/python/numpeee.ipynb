# NumPy Basics

## Array Creation
NumPy arrays, also known as ndarrays (n-dimensional arrays), are the core of the NumPy library. They are used to store collections of homogeneous data (data of the same type). Arrays can be created from Python lists, as shown in the example below:

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
print("Array:\n", arr)
print("Shape:", arr.shape)

c = np.arange(12).reshape(3, 4)
print("\nArange & reshape:\n", c)
```

NumPy arrays have a `shape` attribute that returns a tuple indicating the size of the array in each dimension. You can also create arrays with a sequence of numbers using `np.arange` and reshape them using the `reshape` method.

## Basic Array Operations (Reshaping, Aggregation)
NumPy arrays support various operations, including reshaping and aggregation. Reshaping allows you to change the dimensions of an array without changing its data. Aggregation functions perform calculations on array elements.

Examples of aggregation functions include `np.sum`, `np.mean`, `np.argmax`, and `np.where`. These functions can operate on the entire array or along a specific axis (rows or columns).

```python
import numpy as np

def analyze_marks(marks: np.ndarray):
    """
    Analyze a marks array.

    Parameters
    ----------
    marks : np.ndarray
        2D array where rows are students and columns are subjects.

    Prints
    ------
    - Student-wise averages
    - Subject-wise topper indices
    - Locations and values of marks >= 90
    """
    print("Marks:\n", marks)

    avg = np.mean(marks, axis=1)
    print("\nStudent averages:", avg)

    toppers = np.argmax(marks, axis=0)
    print("Topper index per subject:", toppers)

    x, y = np.where(marks >= 90)
    print("\nMarks >= 90:")
    for i, j in zip(x, y):
        print(f"Row {i}, Col {j}: {marks[i, j]}")


# Example usage
marks = np.array([
    [98, 99, 88],
    [42, 47, 98],
    [56, 49, 72],
    [45, 55, 65],
    [34,100, 45]
])

analyze_marks(marks)
```

# NumPy Performance Advantages

NumPy addresses the efficiency and performance limitations inherent in standard Python lists primarily through **vectorization**, high-speed implementation using C/C++, and built-in optimized functions.

## Vectorization
NumPy converts calculations into a **vectorized format**. This means that operations can be performed on all elements of the array "all at once". NumPy's **vectorized calculations** enable things to happen "in parallel," leading to significant performance gains compared to the element-wise operations required with standard Python lists.

## High-Speed Implementation (C/C++)
NumPy achieves its speed because most of its underlying code is written in **C or C++**. NumPy functions act almost like a wrapper, enabling the user to call Python functions while the high-speed C calculations are executed in the background. This allows NumPy code to be "almost as fast as code that is written in C or C++".

## Broadcasting
NumPy enables an essential operation called **broadcasting**. Broadcasting allows every single element in an array to be influenced by a scalar (like 2 or 1.5) with just "one operation" or calculation. This facilitates rapid element-wise addition or multiplication across the entire array, unlike standard Python lists which duplicate the list when multiplied by a scalar.

## Optimized Mathematical Functions
NumPy offers extensive built-in functions optimized for array operations, replacing the need for complex custom logic. NumPy includes an array of built-in functions for mathematical operations (like `np.sqrt()`, `np.log()`, `np.sum()`, `np.mean()`, and `np.argmax()`) that make calculations easier and are designed to execute quickly. These functions allow users to perform operations like calculating the mean along a specific axis (row or column) in a single, highly optimized command.

# Visualizing XOR with NumPy

This section demonstrates how to generate and visualize data with XOR logic using NumPy and Matplotlib.

## Generating XOR Data
The XOR (exclusive OR) logic returns true if one of the inputs is true but not both. In this example, we generate 200 data points with two features (x and y coordinates) using `np.random.randn`. The labels (`y_xor`) are created by applying the XOR logic to the signs of the two features: `y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)`. The boolean results are then converted to integers (1 for True, 0 for False).

```python
# Generating XOR Data
import numpy as np # Import numpy again for standalone example

np.random.seed(1) # Set the random seed for reproducibility of results
X_xor = np.random.randn(200, 2) # Generate 200 data points with 2 features (x and y coordinates) from a standard normal distribution
# Apply the XOR logic: True if (feature1 > 0) XOR (feature2 > 0)
y_xor = np.logical_xor(X_xor[:, 0] > 0,
                       X_xor[:, 1] > 0)
y_xor = np.where(y_xor, 1, 0) # Convert boolean results to integers (1 for True, 0 for False)

print("First 5 data points (X_xor):")
print(X_xor[:5]) # Print the first 5 generated data points
print("\nFirst 5 labels (y_xor):")
print(y_xor[:5]) # Print the first 5 generated labels
print("\nShape of X_xor:", X_xor.shape) # Print the shape of the feature array
print("Shape of y_xor:", y_xor.shape) # Print the shape of the label array
```

## Plotting the XOR Data
The generated data is visualized using a scatter plot with Matplotlib. Points belonging to Class 1 (where XOR logic is true) are plotted with square markers and 'royalblue' color, while points belonging to Class 0 (where XOR logic is false) are plotted with circle markers and 'tomato' color. This visualization clearly shows that the two classes are not linearly separable, illustrating the need for non-linear classification methods for such datasets.

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_xor_data(n_samples=200, seed=1, plot=True):
    """
    Generate and optionally plot XOR dataset.

    Parameters
    ----------
    n_samples : int
        Number of data points to generate.
    seed : int
        Random seed for reproducibility.
    plot : bool
        If True, displays a scatter plot of the data.

    Returns
    -------
    X : np.ndarray
        Features array of shape (n_samples, 2).
    y : np.ndarray
        Labels array (0 or 1) of shape (n_samples,).
    """
    np.random.seed(seed)
    X = np.random.randn(n_samples, 2)
    y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0)
    y = np.where(y, 1, 0)

    if plot:
        plt.figure(figsize=(6, 6))
        plt.scatter(X[y == 1, 0], X[y == 1, 1],
                    c='royalblue', marker='s', edgecolor='black', s=50, label='Class 1')
        plt.scatter(X[y == 0, 0], X[y == 0, 1],
                    c='tomato', marker='o', edgecolor='black', s=50, label='Class 0')
        plt.xlim([-3, 3]); plt.ylim([-3, 3])
        plt.xlabel('Feature 1'); plt.ylabel('Feature 2')
        plt.title('XOR Data Visualization')
        plt.legend(); plt.grid(True); plt.tight_layout()
        plt.show()

    return X, y


# Example usage
X_xor, y_xor = generate_xor_data()
print("Shapes:", X_xor.shape, y_xor.shape)
```

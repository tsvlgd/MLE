#### plotting only mean

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your DataFrame
mean_petal_length = df['petal_length'].mean()  # Calculate the mean

# Plot the histogram
plt.hist(df['petal_length'], bins=20, alpha=0.7, color='blue', edgecolor='black')

# Add a vertical line at the mean
plt.axvline(mean_petal_length, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_petal_length:.2f}')

# Add labels and legend
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length with Mean')
plt.legend()

# Show the plot
plt.show()
```

```python
# feature relationship
plt.scatter(df['sepal_length'], df['petal_length'], c=df['species'].astype('category').cat.codes)
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.title('feature relationship')
plt.show()
```

```python
# Create a boxplot for petal_length by species
sns.boxplot(x='species', y='petal_length', data=df)

# Add a title and labels
plt.title('Comparison of Petal Lengths by Iris Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')

# Show the plot
plt.show()
```

boxplot compares the petal lengths of three iris species: **setosa**, **versicolor**, and **virginica**. Heres what you can infer, along with answers to your doubts about the dots, width, and separation.

---

#### **1. Central Tendency (Median)**
- The **line inside each box** represents the median (50th percentile) of petal lengths for each species.
  - **Setosa**: Median petal length is around **1.5 cm**.
  - **Versicolor**: Median is around **4.3 cm**.
  - **Virginica**: Median is around **5.5 cm**.

**Inference**: Virginica has the longest median petal length, followed by versicolor, and setosa has the shortest.

---

#### **2. Spread (Interquartile Range, IQR)**
- The **box** represents the interquartile range (IQR), which contains the middle 50% of the data.
  - **Setosa**: IQR is very narrow (~1.3 cm to 1.7 cm), indicating low variability.
  - **Versicolor**: IQR is wider (~4 cm to 5 cm), showing moderate variability.
  - **Virginica**: IQR is the widest (~5 cm to 6.5 cm), indicating high variability.

**Inference**: Virginica has the most variation in petal length, while setosa is the most consistent.

---

#### **3. Range and Outliers (Dots)**
- The **whiskers** extend to the smallest and largest values within 1.5IQR from the quartiles.
- **Dots outside the whiskers** are outliersunusual values that fall outside the expected range.
  - **Setosa**: Has a few outliers below 1.3 cm.
  - **Versicolor**: Has a few outliers below 3 cm.
  - **Virginica**: No visible outliers.

**Inference**: Setosa and versicolor have some petals that are unusually short compared to the rest of their species.

---

#### **4. Width of the Boxes**
- The **width of the boxes** is usually the same for all groups in a boxplot. It does **not** represent the number of data points or any statistical property. Its purely for visual clarity.
- If the widths were proportional, it would typically indicate sample size, but this is not standard in most boxplots.

---

#### **5. Separation Between Groups**
- The **separation between the boxes** (along the x-axis) is purely for visual clarity. It helps distinguish between the species and does not represent any statistical relationship.
- The x-axis is categorical (species names), not numerical.

---

### **Summary of Key Insights**
- **Virginica** has the longest and most variable petal lengths.
- **Setosa** has the shortest and least variable petal lengths.
- **Versicolor** falls in between, with moderate variability.
- **Outliers** are present in setosa and versicolor, indicating some unusually short petals for those species.

```python
sns.violinplot(x="pclass", y="fare", hue="survived", split=True, data=df2)
plt.title("Fare Distribution by Pclass and Survival")
plt.show()
```

Key Observations from the Plot:


Pclass 1:

The fare distribution is highly skewed, with most fares concentrated at the lower end, but some passengers paid very high fares (up to ~500).
The violin is split into two halves (blue for non-survivors, orange for survivors).
Survivors (orange) tend to have a slightly higher concentration of higher fares compared to non-survivors (blue).



Pclass 2:

The fare distribution is more compact, with fares mostly between 0 and 50.
The split violins show that survivors and non-survivors have similar fare distributions, but survivors may have a slight edge in higher fares.



Pclass 3:

The fare distribution is very narrow, with most fares below 50.
The split violins show little difference between survivors and non-survivors in terms of fare distribution.




3. Why Use a Violin Plot?

Shows Distribution Shape: Unlike a boxplot, a violin plot shows the full distribution of the data, including peaks, skewness, and multimodality.
Compares Groups: The split violins allow you to compare the fare distribution between survivors and non-survivors within each passenger class.
Highlights Density: The width of the violin at any point represents the density of data points at that fare value. Wider sections indicate more passengers paid that fare.
Outliers and Spread: The plot also shows the spread and potential outliers in the data, similar to a boxplot.


4. Insights from the Plot

Higher fares in Pclass 1 are associated with higher survival rates, likely because higher fares correlate with better cabins and priority during evacuation.
Pclass 2 and 3 show less difference in fare distribution between survivors and non-survivors, suggesting that fare had less impact on survival in these classes.
The skewness in Pclass 1 indicates that while most passengers paid moderate fares, a few paid extremely high fares, possibly for luxury accommodations.


Summary
This violin plot provides a detailed visual comparison of how fare distribution varies by passenger class and survival status. It highlights that higher fares in 1st class were associated with better survival odds, while the effect is less pronounced in 2nd and 3rd class. The split violins make it easy to see these differences at a glance.

### Pairplot - Multivariate analysis

```python
sns.pairplot(df2[["age", "fare", "pclass", "survived"]].dropna(), hue="survived")
plt.suptitle("Pairplot of Features", y=1.02)
plt.show()
```

Here are the **key insights** from the pairplot in short:

---

### **1. Age vs. Survival**
- **Distribution**: Age is mostly concentrated between 20-40 years for both survivors and non-survivors.
- **Scatter Plot**: Younger passengers (especially children) have a slightly higher survival rate.
- **No strong pattern**: Age alone does not strongly determine survival.

---

### **2. Fare vs. Survival**
- **Distribution**: Most fares are low, but a few passengers paid very high fares.
- **Scatter Plot**: Passengers who paid **higher fares** (especially above 50) tend to have a **higher survival rate**.
- **Outliers**: Some high-fare passengers did not survive, but the trend favors survival for higher fares.

---

### **3. Pclass vs. Survival**
- **Distribution**: Most passengers were in **3rd class**, followed by 1st and 2nd.
- **Scatter Plot**: **1st class passengers** (pclass=1) have a **higher survival rate** compared to 2nd and 3rd class.
- **Clear trend**: Lower pclass (higher class) is strongly associated with survival.

---

### **4. Relationships Between Features**
- **Age vs. Fare**: No clear correlation between age and fare.
- **Age vs. Pclass**: Younger passengers are spread across all classes, but children in higher classes may have better survival.
- **Fare vs. Pclass**: Higher fares are associated with **1st class**, which aligns with the higher survival rate for 1st class passengers.

---

### **5. Summary of Key Insights**
- **Higher fare and lower pclass (1st class) passengers had better survival rates**.
- **Age alone is not a strong predictor of survival**, but younger passengers (especially children) had a slight advantage.
- **Pclass is the strongest predictor of survival**, with 1st class passengers surviving the most.

### EDA on titanic Dataset

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
df = sns.load_dataset('titanic')
df.tail()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Drop less useful or high-missing columns
df.drop(columns=['deck', 'embark_town'], inplace=True)

# Fill missing age with median
df['age'].fillna(df['age'].median(), inplace=True)

# Fill embarked with mode
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)

# Drop rows with any remaining missing values
df.dropna(inplace=True)

# Confirm cleanup
df.isnull().sum()
```

#### Univariate analysis -> (Identify outliers, skewness, or patterns in a single variable.)

```python
sns.countplot(x='survived', data=df)
plt.xlabel("Survived (0=No, 1=Yes)")
plt.ylabel('No of passengers')
plt.title("Survival Count")
```

#### **Age** Distribution

```python
sns.histplot (df['age'], kde=True, bins=20)
plt.xlabel("Age")
plt.ylabel('Count')
plt.title("Age Distribution")
plt.show()
```

```python
df.describe()
```

```python
sns.countplot(x='pclass', data=df)
plt.xlabel("Pclass")
plt.ylabel('No of passengers')
plt.title("Survival Count")
plt.show()
```

#### Bivariate analysis -> (eg: among the male how often you survive )

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.title('Survival By Gender')
plt.show()

# Proof: Women always live longer than men.
```
